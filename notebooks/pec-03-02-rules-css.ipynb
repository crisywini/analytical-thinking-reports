Market Basket Analysis

Dataset

```python
data_arff= arff.loadarff('../data/raw/supermarket.arff')
data = pd.DataFrame(data_arff[0])
```

```python
data
```

```python
possible_values = list(data['department1'].unique())
possible_values
```

```python
for e in data:
    print(e)
    data[e] = data[e].replace(possible_values[0], False)
    data[e] = data[e].replace(possible_values[1], True)
```

```python
unique_types = list(data['total'].unique())
```

```python
data['total'] = np.array(map(lambda x: "high" if x == unique_types[0]  else "low", data['total']))
```

```python
X = data.drop('total', axis=1)
y = data['total']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=12345) 
```

```python
X_train
```

```python
y_train.unique()
```

Using Ripper rules

```python
import wittgenstein as lw
ripper_classifier = lw.RIPPER()
ripper_classifier.fit(X_train, y_train, pos_class='high', random_state=2)
```

```python
precision = ripper_classifier.score(X_test, y_test, precision_score)
recall = ripper_classifier.score(X_test, y_test, recall_score)
cond_count = ripper_classifier.ruleset_.count_conds()

print(f'precision: {precision} recall: {recall} conds: {cond_count}')
```

```python
ripper_classifier.ruleset_.out_pretty()
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {"prune_size": [0.33, 0.5, 0.2], "k": [1, 2]} 
grid = GridSearchCV(estimator=lw.RIPPER(), param_grid=param_grid)
grid.fit(X_train, y_train, pos_class='high')  
```

```python
grid.best_params_
```

```python
ripper_classifier = lw.RIPPER()
ripper_classifier.fit(X_train, y_train, pos_class='high', k= 1, prune_size= 0.5)
```

```python
precision = ripper_classifier.score(X_test, y_test, precision_score)
recall = ripper_classifier.score(X_test, y_test, recall_score)
cond_count = ripper_classifier.ruleset_.count_conds()

print(f'precision: {precision} recall: {recall} conds: {cond_count}')
```
